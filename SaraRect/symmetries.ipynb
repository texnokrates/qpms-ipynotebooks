```python
from qpms import symmetries
import numpy as np
import qpms
```

```python
from qpms.symmetries import Permutation, SVWFPointGroupInfo, point_group_info
point_group_info['x_and_z_flip_group'] =   SVWFPointGroupInfo(
    'x_and_z_flip_group',
    (
        Permutation(0,1, size=4), # x -> -x mirror op
        Permutation(2,3, size=4), # z -> -z mirror op
    ),
    {
        "P'": (1, 1),
        "R'": (-1, 1),
        "P''": (-1,-1),
        "R''": (1, -1),
    },
    lambda lMax : (qpms.xflip_tyty(lMax), qpms.zflip_tyty(lMax))
)
point_group_info['y_and_z_flip_group'] =   SVWFPointGroupInfo(
    'y_and_z_flip_group',
    (
        Permutation(0,1, size=4), # y -> -y mirror op
        Permutation(2,3, size=4), # z -> -z mirror op
    ),
    {
        "P'": (1, 1),
        "R'": (-1, 1),
        "P''": (-1,-1),
        "R''": (1, -1),
    },
    lambda lMax : (qpms.yflip_tyty(lMax), qpms.zflip_tyty(lMax))
)
```

```python
group_info = symmetries.point_group_info
```

```python
lMax = 3
for group, info in group_info.items():
    print ('\n=======',group,'=======')
    projs = info.svwf_irrep_projectors(lMax, 'x') if group == 'D3h' else info.svwf_irrep_projectors(lMax)
    for rep, ar in projs.items():
        print('_______',rep,'_______')
        for i in np.ndindex(ar.shape):
            if (ar[i] != 0):
                print(i, ar[i])
```
