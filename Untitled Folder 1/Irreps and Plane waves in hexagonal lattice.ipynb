```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import math
ň = None
s3 = math.sqrt(3)
pp = np.pi/2 #pípůl
pt = np.pi/3 #pítřeť
```

```python
def kpointplanewaves(phases = (0,0,0,0,0,0), facs = (1,1,1,1,1,1), Kmultiplier=1, ascale=40, wscale=25, sscale=100, gphase=0, showcontribs=False, noplot=False):
    facs = np.array(facs)
    phases = np.array(phases,dtype=float)
    phases += gphase
    hsl = 1 # hexagon side length
    n1 = 5
    n2 = 5
    b1 = np.array((0,hsl*s3))
    b2 = np.array((hsl*3/2,hsl*s3/2))
    Ashift = np.array((-hsl,0))
    Bshift = np.array((hsl,0))

    Asites = np.arange(-n1,n1)[:,ň,ň] * b1[ň,ň,:] + np.arange(-n2,n2)[ň,:,ň] * b2[ň,ň,:] + Ashift
    Asites = np.reshape(Asites,(np.prod(Asites.shape[:-1]), Asites.shape[-1]))
    Bsites = Asites - Ashift + Bshift

    Klen = 4*np.pi/3/s3/hsl*Kmultiplier
    wi = np.arange(6)
    K = Klen * np.stack((np.cos(np.pi*(1/6+wi/3)),np.sin(np.pi*(1/6+wi/3))), axis=-1)
    up = np.stack((np.cos(np.pi*(4/6+wi/3)),np.sin(np.pi*(4/6+wi/3))), axis=-1)
    if not noplot:
        f, ax = plt.subplots(1)
    else:
        f = None
        ax = None
    
    if not noplot:
        for i in range(3): # osy
            ax.plot(*np.transpose([-4*K[i],4*K[i]]),ls='-',c='k')
    iticks = np.arange(-32*n1,32*n1)[ň,:,ň]*K[:,ň,:]  / Klen * s3/32 # [wi, site, x]
    iphases = np.sum(iticks * K[:,ň,:], axis=-1 )-phases[:,ň] # [wi, site, /x/] * [wi, ň,/x/] - [wi, ň,] ~ [wi, site]
    phaseA = np.tensordot(K, Asites, axes=(-1,-1))-phases[:,ň] # ([wi,/x/]*[site,/x/])-[wi, ň] ~ [wi,site]
    phaseB = np.tensordot(K, Bsites, axes=(-1,-1))-phases[:,ň]
    iwaves = np.cos(iphases)[:, :, ň] * up[:, ň, :] * facs[:, ň, ň] # cos[wi,site,ň]*[wi,ň,x]*[wi,ň,ň] ~ [wi,site,x]
    Awave = np.sum(np.cos(phaseA)[:, :, ň] * up[:, ň, :] * facs[:, ň, ň], axis=0) # [/wi/,site,ň]*[/wi/,ň,x]*[/wi/,ň,ň] ~ [site,x]
    AwaveH = np.sum(np.sin(phaseA)[:, :] * facs[:, ň], axis=0)
    Awave*=(abs(Awave)>1e-13)
    Bwave = np.sum(np.cos(phaseB)[:, :, ň] * up[:, ň, :] * facs[:, ň, ň], axis=0)
    BwaveH = np.sum(np.sin(phaseB)[:, :] * facs[:, ň], axis=0)
    Bwave*=(abs(Bwave)>1e-13)
    scale_units = 'width'
    
    if not noplot:
        for i in range(6):
            ax.quiver(iticks[i,:,0], iticks[i,:,1],
                       iwaves[i,:,0],
                       iwaves[i,:,1],
                     lw=0.2, color= 'green' if (i < 3) else 'red', headwidth=0, scale_units=scale_units, scale = wscale)

        sites = np.concatenate((Asites, Bsites))
        waveH = np.concatenate((AwaveH, BwaveH))
        #print (abs(waveH)*sscale)
        ax.scatter(Asites[:,0], Asites[:,1], marker='+', c='orange')
        ax.scatter(Bsites[:,0], Bsites[:,1], marker='+', c='magenta')
        ax.scatter(sites[:,0], sites[:,1], marker='o', s=abs(waveH)*sscale,
                   c=np.where(waveH>0,'pink','lightblue') )
        ax.quiver(Asites[:,0],Asites[:,1],Awave[:,0],Awave[:,1],headwidth=1, scale_units = scale_units, scale=ascale)
            
        ax.quiver(Asites[:,0],Asites[:,1],Awave[:,0],Awave[:,1],headwidth=1, scale_units = scale_units, scale=ascale)
        ax.quiver(Bsites[:,0],Bsites[:,1],Bwave[:,0],Bwave[:,1],headwidth=1, scale_units = scale_units, scale=ascale)
        if showcontribs:
            for i in range(6):
                ax.quiver(Asites[:,0],Asites[:,1],(np.cos(phaseA)[:, :, ň] * up[:, ň, :] * facs[:, ň, ň])[i,:,0],(np.cos(phaseA)[:, :, ň] * up[:, ň, :] * facs[:, ň, ň])[i,:,1],headwidth=1, scale_units = scale_units, scale=ascale, color='green' if (i < 3) else 'red')
                ax.quiver(Bsites[:,0],Bsites[:,1],(np.cos(phaseB)[:, :, ň] * up[:, ň, :] * facs[:, ň, ň])[i,:,0],(np.cos(phaseB)[:, :, ň] * up[:, ň, :] * facs[:, ň, ň])[i,:,1],headwidth=1, scale_units = scale_units, scale=ascale, color='green' if (i < 3) else 'red')

        #plt.plot(*np.transpose(upiticks),ls='',marker='o')
        #plt.plot(*np.transpose(rupiticks),ls='',marker='o')
        #plt.plot(*np.transpose(lupiticks),ls='',marker='o')
        ax.set_xlim(-4.5,4.5)
        ax.set_ylim(-3,3)
        ax.set_title('%.3f π' % (gphase/np.pi))
    return (Awave,Bwave,AwaveH, BwaveH, Asites, Bsites, iticks, iwaves, f,  ax)

def justplot(Awave, Bwave, AwaveH, BwaveH, Asites, Bsites, iticks, iwaves, ascale=20, wscale=20, sscale=100):
    hsl=1
    Klen = 4*np.pi/3/s3/hsl#*Kmultiplier
    wi = np.arange(6)
    K = Klen * np.stack((np.cos(np.pi*(1/6+wi/3)),np.sin(np.pi*(1/6+wi/3))), axis=-1)
    up = np.stack((np.cos(np.pi*(4/6+wi/3)),np.sin(np.pi*(4/6+wi/3))), axis=-1)

    
    sites = np.concatenate((Asites, Bsites))
    waveH = np.concatenate((AwaveH, BwaveH))

    f, ax = plt.subplots(1)
    for i in range(3): # osy
        ax.plot(*np.transpose([-4*K[i],4*K[i]]),ls='-',c='k')

    scale_units = 'width'
    for i in range(6):
        ax.quiver(iticks[i,:,0], iticks[i,:,1],
                   iwaves[i,:,0],
                   iwaves[i,:,1],
                 lw=0.2, color= 'green' if (i < 3) else 'red', headwidth=1, scale_units=scale_units, scale = wscale)
    sites = np.concatenate((Asites, Bsites))
    waveH = np.concatenate((AwaveH, BwaveH))
    ax.scatter(Asites[:,0], Asites[:,1], marker='+', c='orange')
    ax.scatter(Bsites[:,0], Bsites[:,1], marker='+', c='magenta')
    ax.scatter(sites[:,0], sites[:,1], marker='o', s=abs(waveH)*sscale,
               c=np.where(waveH>0,'pink','lightblue') )
    ax.quiver(Asites[:,0],Asites[:,1],Awave[:,0],Awave[:,1],headwidth=0, scale_units = scale_units, scale=ascale)
    ax.quiver(Bsites[:,0],Bsites[:,1],Bwave[:,0],Bwave[:,1],headwidth=0, scale_units = scale_units, scale=ascale)
    ax.set_xlim(-4.5,4.5)
    ax.set_ylim(-3,3)
    return (f, ax)
```
